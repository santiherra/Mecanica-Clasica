### Ejercicio 2: tiempo de caída de una bola.

 Una bola se deja caer desde una torre de altura $h$ con velocidad inicial cero y cae acelerada hacia abajo por la fuerza de la gravedad $g$.

Escribir un programa que pida al usuario que introduzca la altura $h$ de la torre en metros y calcule e imprima el tiempo que tarda la bola en llegar al suelo, ignorando la resistencia del aire. Usar el programa para calcular el tiempo que tarda en llegar al suelo una bola que se lanza desde una torre de 100 metros de altura.

#### Solución

La bola debe recorrer una distancia igual a la altura de la torre:
$$h = \frac{1}{2} g t^2. $$
Despejando $t$ se obtiene:
$$ t = \sqrt{\frac{2 h}{g} }.$$

```python
from math import sqrt
g = 9.81
h = float(input("Introduzca la altura de la torre en metros:"))
t = sqrt(2*h/g)
print("El tiempo que tarda la bola en golpear el suelo es", t, "segundos")
```

### Ejercicio 4: cambio de coordenadas polares a cartesianas.

Escribir un progama que tome las coordenas polares ($r$, $\theta$) de un punto del plano introducidas por el usuario y calcule e imprima sus coordenadas cartesianas.

Pasos:

  1. Obtener del usuario los valores de $r$ y $\theta$.
  2. Convertir las coordenadas polares a cartesianas  ($x$, $y$) usando las relaciones:
    
  $$ x = r \cos \theta, \qquad y = r \sin \theta$$
  3. Imprimir los resultados.
  
#### Solución

```python
from math import sin,cos,pi
r = float(input("Introduzca r: "))
d = float(input("Introduzca theta en grados: "))

theta = d*pi/180
x,y = r*cos(theta),r*sin(theta)

print("x =", x, ", y =",y)
```

### Ejercicio 5: cambio de coordenadas cartesianas a polares planas.

Escribir un programa que realice el cambio de coordenadas cartesianas a coordenadas polares en el plano.



#### Solución:

$$ r = \sqrt{x^2+y^2},\qquad \theta = \arctan\left(\frac{y}{x}\right).$$

* Notas: la función arcotangente es multivaluada. La función atan depende un argumento y atan2 de dos.

```python
from math import sqrt,atan2,pi

#preguntamos al usuario por los valores de x e y.
x = float(input("introduzca el valor de x: "))
y = float(input("introduzca el valor de y: "))

# convertimos a coordenadas polares
r,theta = sqrt(x**2 + y**2), atan2(y,x)*180/pi # theta en grados

# Imprimimos los resultados
print(" r = ", r, ", theta = ", theta)
```

### Ejercicio 6: un viaje espacial.

Un cohete viaja en línea recta a una velocidad relativista $v$ desde la Tierra a otro planeta a una distancia de $x$ años luz. Escribir un programa que pida el valor de $x$ y el de $v/c$, y calcule el tiempo en años que mide:

1. Un observador en reposo en la Tierra.

2. Un pasajero a bordo del cohete.

Usar el programa para el caso de un planeta a 10 años luz de la Tierra cuando $v/c = 0.99$.


*Ayuda:*
la contracción de Lorentz–Fitzgerald reduce la distancia al destino para el observador en el cohete, y por tanto el tiempo de viaje, por un factor $$\gamma = \frac{1}{\sqrt{1-\frac{v^2}{c^2}}}.$$

```python
from math import sqrt

x = float(input("Introduzca la distancia al planeta en años luz: "))
v = float(input("Introduzca la velocidad del cohete como fracción de c: "))

gamma = 1/sqrt(1-v**2)

print("El tiempo medido en la Tierra es", x/v, "años")
print("El tiempo medido en el cohete es", x/(v*gamma), "años")
```

## La instrucción "if"

* Queremos que un progama haga algo únicamente si se cumple una cierta condición.

Ejemplos:

```python
x = int(input("Introduzca un número entero que no sea mayor que diez: "))
if x>10:
    print(" Usted introdujo un número mayor que diez.")
    print(" Déjeme arreglar esto por usted.")
    x=10
print("Su número es", x)
```

Notas:

```python
x = int(input("Introduzca un número entero: "))
if x>10:
    print("x>10")
else:
    print(" x es menor ó igual a 10")   
```

## La instrucción "while".

*El bucle "while" ejecuta el bloque de órdenes indentado si se cumple la condición y después vuelve a la instrucción while. Si se cumple la condición vuelve a ejecutar las instrucciones indentadas y continúa así hasta que deje de cumplirse la condición.*

PELIGRO: ¡podemos caer en un bucle infinito!

Veamos un ejemplo para explicarlo.

```python
x = int(input("Introduzca un número entero que no sea mayor que diez: "))
while x>10:
    print("Este número es mayor que diez, por favor pruebe otra vez.")
    x = int(input("Introduzca un número entero que no sea mayor que diez: "))
print("Su número es el", x)
```

## Las instrucciones "break" y "continue"

La instrucción "break" permite salir de un bucle while si se cumple una condición determinada.

Veamos un ejemplo:


```python
x = int(input("Introduzca un número entero que no sea mayor que diez: "))
while x>10:
    print("Este número es mayor que diez, por favor, pruebe otra vez.")
    x = int(input("Introduzca un número entero que no sea mayor que diez o acierte el número mágico: "))
    if x==111:
        break
print("Su número es el", x,", que o no es mayor que 10 o es el número mágico 111")
```

**Nota:** si el bucle  "while" está seguido de una instrucción  "else", la instrucción "else" no se ejecuta despues del break. Esto permite crear un programa que hace cosas diferentes si el bucle "while" finaliza normalmente o via break.

$$\bar{x} = \left[ \prod_{i=1}^n x_i\right]^{\frac{1}{n}}$$
$$\ln \bar{x} = \frac{1}{n} \sum_{i=1}^n \ln x_i$$
$$ \bar{x} = \exp \left( \frac{1}{n} \sum_{i=1}^n \ln x_i \right)$$

```python
from numpy import loadtxt
from math import log,exp
valores = loadtxt("valores.txt",float)
print(valores)
logs = array(list(map(log,valores)),float)
print(logs)
geometric = exp(sum(logs)/len(logs))
print(geometric)
```

### Ejercicio 12: líneas de emisión del espectro del átomo de hidrogeno.


Calcular las longitudes de onda $\lambda$ de las líneas de emisión del átomo de hidrogeno:

$$\frac{1}{\lambda}= R \left( \frac{1}{m^2}-\frac{1}{n^2}\right),$$

donde $R= 1.097 \times 10^{-2}\: {\rm nm}^{-1}$ es la llamada constante de Rydbergy $m$ y $n$ son número enteros positivos.

Para un valor dado de $m$, las longitudes de onda $\lambda$ obtenidas mediante esta fórmula para todos los valores de $ n > m$ forman series. Las tres primeras, $ m = 1, 2, 3$, se conocen como series de *Lyman*, *Balmer* y *Paschen*. Veamos un programa de Python que calcula e imprime los primeros 5 terminos de estas series.


```python
R = 1.097e-2
for m in range(1,4):
    print("Series para m=", m)
    for k in range(1,6):
        n = m + k 
        invlambda = R*(1/m**2-1/n**2)
        print(" ", 1/invlambda, "nm")
    print("...")
    print()
    
```

## Variables locales

* En el ejemplo anterior, $f$ y $k$ son variables locales. 
* Las variables creadas en el interior de la definición de una función existen sólo dentro de la función. Son por tanto variables *locales*.
* No se puede hacer referencia a ellas fuera del programa.
* Nótese que lo contrario sí se puede hacer: se puede llamar una variable global definida fuera de la función en el interior de la función.

### Ejercicio 13: definir una función que calcule los factores primos de un entero $n$

```python
def factores(n):
    listadefactores = []
    k = 2 
    while k<=n:
        while n%k==0:
            listadefactores.append(k)
            n //= k
        k +=1
    return listadefactores
            
```

Ejemplo de programa:  
    

```python
def primos(max_primo):
    lista_primos = [2]
    n = 1
    while 2*n+1 <= max_primo: # números pares mayores que 2 no pueden ser primos
        k = 1 
        while (2*k+1)**2 <= 2*n+1:
            if (2*n+1)%(2*k+1)==0:
                n += 1
                break
            k += 1
        else:
            lista_primos.append(2*n+1)
            n += 1
    return(lista_primos)

        
```

### Ejercicio 15: la fórmula semiempírica de la masa.

En física nuclear, la fórmula semiempírica de masa es una ecuación que permite calcular de forma aproximada la energía de ligadura nuclear $B$ de un nucleo atómico con número atómico $Z$ y número másico $A$ (aproximadamente igual al número de protones más neutrones del núcleo):

$$ B = a_1 A - a_2 A^{2/3} - a_3 \frac{Z^2}{A^{1/3}} - a_4 \frac{(A - 2 Z)^2}{A} + \frac{a_5}{A^{1/2}} ,$$

donde, en unidades de MeV (= millones de electron-voltios), las constantes son:

$a_1 = 15.8$, $a_2 = 18.3$, $a_3 = 0.714$, $a_4 = 23.2$, y 

$$ a_5 = \begin{cases} 0 & \text{si } A \text{ es impar,}\\
                 12.0 & \text{si } A \text{ y } Z \text{ son ambos pares,} \\
                 -12.0 & \text{si } A \text{ es par y } Z \text{ es impar.}
        \end{cases}$$

1. Escribir un programa que tome como entrada los valores de $A$ y $Z$ e imprima la energía de ligadura para el átomo correspondiente. Usar el programa para calcular la energía de ligadura de un átomo con $A = 58$ y $Z=28$. (La respuesta correcta es aproximadamente 500MeV.)

2. Modificar el programa para que imprima no la energía total de ligadura $B$, sino la energía de ligadura por nucleón, es decir, $B/A$.

3. Modificar el programa para que tome como entrada solamente el valor del número atómico $Z$ y recorra todos los valores de $A$ desde $A=Z$ hasta $A= 3 Z$, para encontrar cuál de los valores de $A$ corresponde a la mayor energía de ligadura por nucleón. Hacer que el programa imprima el valor de $A$ para el núcleo mas estable y el valor de la energía de ligadura por nucleón.

4. Modificar de nuevo el programa para que en vez de tomar $Z$ como entrada, recorra todos los valores de $Z$ desde 1 hasta 100 e imprima el valor más estable de $A$ para cada valor de $Z$. ¿Para qué valor de de $Z$ la energía por nucleón es máxima? (La respuesta correcta, en la vida real, es $Z=28$, que corresponde al níquel).

```python
# Programa de la fórmula semi-empírica de masa

# Definimos los valores de las constantes.

a1 = 15.8
a2 = 18.3
a3 = 0.714
a4 = 23.2

# Calculamos la energía por nucleón.

def por_nucleon(Z,A):
    
    # Calculamos la constante a5
    
    if A%2 == 1:
        a5 = 0.0
    elif Z%2 == 0:
        a5 = 12.0
    else:
        a5 = -12.0
        
    # Calculamos la energía de ligadura B
    B = a1*A - a2*(A**(2/3))-a3*(Z**2)/(A**(1/3)) - a4*((A - 2*Z)**2)/A + a5/(A**(1/2))
    return B/A # energía de ligadura por nucleón

# Programa principal

nlist=[]
for Z in range (1,101):
    maxbpn = 0.0
    for A in range(Z,3*Z+1):
        bpn = por_nucleon(Z,A)
        if bpn > maxbpn:
            maxbpn = bpn
            maxA = A
    nlist.append([Z,maxA,maxbpn]) 
    # en la lista nlist se almacena el valor óptimo de A y la energía de ligadura por nucleón B/A para cada Z 
    print("Z =",Z,", A =", maxA, ", B/A =", maxbpn)
    
            
```

```python
from operator import itemgetter
print(sorted(nlist,key=itemgetter(2),reverse=True)) # imprime la lista de núcleos por orden descendente de B/A
```

Al ejecutar el programa se comprueba que el máximo valor de B/A ocurre para Z = 28 y A = 62, siendo B/A = 8.70245768367189 MeV. (El valor experimental de B/A para el níquel es de 8.7945 MeV.)

### Ejercicio 17: interferencia de ondas circulares en un estanque.

Interferencia de dos ondas sinusoidales esféricas:


$$\xi (x,y) = \xi_0 \sin (k r_1) +  \xi_0 \sin (k r_2)$$

$$r_i = \sqrt{(x-x_i)^2+(y-y_i)^2} $$

$$\xi_i (x,y) = \xi_0 \sin (k r_i) $$

donde $i= 1,2$, $k = 2 \pi/\lambda$. Tomar como amplitud $\xi_0 = 1.0$, separación entre los centros de las ondas $20$ cm y longitud de onda $\lambda = 5$ cm. Suponer que el estanque es un cuadrado de $100$ cm de lado y usar una malla de 500x500 puntos.

```python
from math import sqrt,sin,pi
from numpy import empty
from pylab import imshow, show

longitud_onda = 5.0
k = 2*pi/longitud_onda
xi0=1.0
separacion=20.0            # separación de los centros en centímetros
lado=100.0                 # lado del cuadrado en centímetros
puntos=500                 # numero de particiones en cada lado
espaciado=lado/puntos      # espaciado de puntos en cm.

# Calculamos la posicion de los centros de los círculos
x1 = lado/2 + separacion/2
y1 = lado/2
x2 = lado/2 - separacion/2
y2 = lado/2

# Creamos un array para almacenar las alturas del agua en el estanque
xi = empty([puntos,puntos],float)

# calculamos los valores del array
for i in range(puntos):
    x = espaciado*i
    for j in range(puntos):
        y = espaciado*j
        r1 = sqrt((x-x1)**2 + (y-y1)**2)
        r2 = sqrt((x-x2)**2 + (y-y2)**2)
        xi[j,i] = xi0*sin(k*r1) + xi0*sin(k*r2) # recordar que el índice [j,i] corresponde al punto (x,y) de la gráfica
        
# hagamos un gráfico
imshow(xi,origin="lower", extent=[0,lado,0,lado])
show()
```

### Ejercicio 19: distribución asintótica de los números primos.

Según un célebre resultado probado a finales del siglo XIX, el número $\pi(x)$ de números primos menores o iguales que un número real $x>0$ se comporta como $x/\log x$ para $x$ muy grande. Más precisamente, aunque la diferencia $\pi(x)-x/\log x$ tiende a infinito cuando $x\to\infty$, el cociente entre ambas funciones tiende a $1$:
$$ \lim_{x\to\infty}\frac{\pi(x)}{x/\log x}=1.$$
La convergencia de $x/\log x$ a $\pi(x)$ es, sin embargo, bastante lenta.

1. Escribir un programa que calcule $\pi(x)$ para $x$ entre $1$ y un número dado y represente gráficamente dicha función junto con su aproximación $x/\log x$.

2. Representar en otra gráfica el cociente entre $\pi(x)$ y $x/\log x$.

#### Ejemplo de programa:

```python
from pylab import plot,show
from numpy import array,ones
from math import log

def prime_app(x):
    return x/log(x)

max_primo = 10**6
num_primos = [0,1,2]

n = 2
while 2*n+1 <= max_primo: # números pares mayores que 2 no pueden ser primos
    num_primos.append(num_primos[2*n-2])
    k = 1
    while (2*k+1)**2 <= 2*n+1:
        if (2*n+1)%(2*k+1)==0:
            num_primos.append(num_primos[2*n-2])
            n += 1
            break
        k += 1
    else:
        num_primos.append(num_primos[2*n-2]+1)
        n += 1

plot(num_primos,"blue")
plot(range(2,max_primo),list(map(prime_app,range(2,max_primo))),"red")
show() 
del(num_primos[0])
plot(array(num_primos,float)/list(map(prime_app,range(2,max_primo))),"green")
plot(ones(max_primo-2),"blue",linestyle='dashed')
show()
print(num_primos[-1]/prime_app(max_primo-1))
```

### Ejercicio 20: representación de curvas parametrizadas.

1. Hacer un gráfico de la deltoide, que es la curva plana definida paramétricamente como sigue:

$$ x= 2 \cos\theta + \cos (2 \theta),\qquad y = 2 \sin\theta - \sin (2 \theta), $$

donde $0 \leq \theta < 2 \pi$.

*Ayuda:* tomar valores para $\theta$ entre $0$ y $ 2 \pi$ y calcular $x$ e $y$.

2. Represenatr las siguientes curvas plana definidas en coordenadas polares:

  *Espiral galileana:* $\quad r = \theta^2,\enspace 0 \leq \theta \leq 10 \pi$.
   
  *Curva de Fey:* $\quad r = e^{\cos \theta} - 2 \cos (4 \theta) + \sin^5(\frac{\theta}{12})$

```python
from numpy import arange, cos,sin,pi
from pylab import plot,show

pasos = 100
theta = arange(0,2*pi, 2*pi/pasos)
x = 2*cos(theta) + cos(2*theta)
y = 2*sin(theta) - sin(2*theta)
          
plot(x,y,"blue")
show()
```
